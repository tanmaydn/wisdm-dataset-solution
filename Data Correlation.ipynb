```python
accel_phone = np.array(accel_phone)
gyro_phone = np.array(gyro_phone)
```

```python
phone_data = np.concatenate((accel_phone, gyro_phone), axis=1)
phone_data
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=0)
```

```python
y_pred = kmeans.fit_predict(phone_data)
y_pred.shape
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d 
```

```python
ax = plt.axes(projection ="3d") 
  
# Creating plot 
ax.scatter3D(y_pred, phone_data[:,0], phone_data[:,1]) 
ax.scatter3D(y_pred, phone_data[:,3], phone_data[:,4])
# show plot 
plt.show() 
```

```python
plt.scatter(y_pred, phone_data[:,0], color='yellow')
plt.scatter(y_pred, phone_data[:,3], color='red')

centers = np.array(kmeans.cluster_centers_)
plt.scatter(centers[:,0], centers[:,1], marker="x", color='blue')
```
