# **DATA PREPROCESSING**

Here, we would be preprocessing the data and would be considering only the functions and the values which would help us in the analysis segment and would ignore the rest of the data. Either we would drop the data column or we would be simply not consider in the loaded dataset

```python
columns=['user','activity','timestamp','x_ac_ph','y_ac_ph','z_ac_ph']

accel_phone = pd.DataFrame(data=None,columns=columns)
for dirname, _, filenames in os.walk('/kaggle/input/smartphone-and-smartwatch-activity-and-biometrics/wisdm-dataset/wisdm-dataset/raw/phone/accel'):
    for filename in filenames:
        df = pd.read_csv('/kaggle/input/smartphone-and-smartwatch-activity-and-biometrics/wisdm-dataset/wisdm-dataset/raw/phone/accel/'+filename , sep=",", header=None)
        temp=pd.DataFrame(data=df.values, columns=columns)
        accel_phone=pd.concat([accel_phone,temp])
```

# **APPLICATION OF MULTIVARIATE TIME-SERIES ANALYSIS**

Here, we would be using Multivariate time series analysis and would be using a defined window function for creating the window of analysis and then perform the analysis based on the output of the windowing and hence compare with the validation set

```python
train = df_phone_a[:2400]
test = df_phone_a[2401:len(df_phone_a)]
```

```python
train = train.values
test = test.values
```

Here, we are generating the 'xtrain' and the 'ytrain' dataset for feeding into the system of the model. We are manually creating the dataset which would be fed into our model. The dataset has been wholly converted into a numpy array system and hence hence we don't have to deal now with the dataframes.

```python
xtrain = []
ytrain = []

for k in range(len(train)-4):
    xtrain.append(train[k])
    ytrain.append(train[k+4])
```

```python
xtest = []
ytest = []
for j in range(len(test)-2):
    xtest.append(test[j])
    ytest.append(test[j+2])
```

```python
xtrain = np.array(xtrain)
ytrain = np.array(ytrain)
xtest = np.array(xtest)
ytest = np.array(ytest)
```

```python
xtrain.shape
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Activation, Dense, Dropout
```

```python
model = Sequential()
model.add(Dense(512, input_shape=(2396,6)))
model.add(Activation('relu'))
model.add(Dense(6))
model.add(Activation('softmax'))
model.summary()

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
num_epochs =50
batch_size = 128
history = model.fit(xtrain, ytrain,
                    batch_size=batch_size,
                    epochs=num_epochs,
                    verbose=2,
                    validation_split=0.2)
```

```python
ypred = model.predict(xtest)
```

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
```

```python
sqr_err = mse(ytest, ypred)
abs_err = mae(ytest, ypred)
```

```python
## This is the error values for the work 'A' in 'phone'
print(sqr_err)
print(abs_err)
```
